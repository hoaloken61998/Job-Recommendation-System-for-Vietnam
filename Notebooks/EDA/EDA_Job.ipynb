# Phân tích đơn biến
## 1. Phân tích Top địa điểm có nhiều việc làm nhất

```python
top_places = df['Location'].value_counts().nlargest(15)
fig1 = px.bar(
        top_places, y=top_places.index, x=top_places.values,
        orientation='h',
        title="<b>Top 15 Tỉnh/Thành phố có nhiều tin tuyển dụng nhất</b>",
        labels={'y': 'Tỉnh/Thành phố', 'x': 'Số lượng tin đăng'},
        text=top_places.values
    )
fig1.update_layout(yaxis={'categoryorder':'total ascending'})
fig1.show()
```

## 2. Phân tích Top ngành nghề

```python
all_industries = df['Industry'].str.split(', ').explode().str.strip()
top_industries = all_industries.value_counts().nlargest(15)
fig2 = px.bar(
        top_industries, y=top_industries.index, x=top_industries.values,
        orientation='h',
        title="<b>Top 15 Ngành nghề có nhu cầu tuyển dụng cao nhất</b>",
        labels={'y': 'Ngành nghề', 'x': 'Số lượng tin đăng'},
        text=top_industries.values
    )
fig2.update_layout(yaxis={'categoryorder':'total ascending'})
fig2.show()
```

## 3. Phân tích mức lượng trung bình theo Công Việc

```python
# Create a copy of the DataFrame for manipulation
df_salary = df.copy()

# Extract numeric salary values (min and max) for parsing
def extract_salary_range(salary_str):
    if isinstance(salary_str, str):
        nums = re.findall(r'\d[\d,.]*', salary_str.replace(',', ''))
        nums = list(map(int, nums))
        if len(nums) == 2:
            return nums[0], nums[1]
        elif len(nums) == 1:
            return nums[0], nums[0]
    return None, None

df_salary[['Salary_Min', 'Salary_Max']] = df_salary['Salary'].apply(lambda x: pd.Series(extract_salary_range(x)))

# Calculate average salary
df_salary['Salary_Avg'] = df_salary[['Salary_Min', 'Salary_Max']].mean(axis=1)

# Filter out unrealistic values (e.g., >200 million VND)
df_salary_filtered = df_salary[df_salary['Salary_Avg'] < 200_000_000]

salary_by_job = df_salary_filtered.groupby('Job_Name')['Salary_Max'].mean().nlargest(15).sort_values()
fig4 = px.bar(
        salary_by_job, y=salary_by_job.index, x=salary_by_job.values,
        orientation='h',
        title="<b>Top 15 Tỉnh/Thành phố có mức lương trung bình cao nhất</b>",
        labels={'y': 'Tỉnh/Thành phố', 'x': 'Lương trung bình (triệu VND)'},
        text=salary_by_job.values.round(1)
    )
fig4.update_layout(yaxis={'categoryorder':'total ascending'})
fig4.show()
```

# **Phân tích đa biến**


```python
# Chuẩn hóa tên cột về viết thường
df.columns = df.columns.str.lower()

# Xử lý dữ liệu văn bản: chuẩn hóa job_name, location, job_type, skills
df['job_name'] = df['job_name'].astype(str).str.strip().str.lower().str.title()
df['location'] = df['location'].astype(str).str.strip().str.title()
if 'job_type' in df.columns:
    df['job_type'] = df['job_type'].astype(str).str.strip().str.title()
if 'skills' in df.columns:
    df['skills'] = df['skills'].astype(str).str.strip().str.lower()

# 1. Job Title vs Location
top_job_titles = df['job_name'].value_counts().nlargest(10).index
top_locations = df['location'].value_counts().nlargest(10).index

filtered_df = df[df['job_name'].isin(top_job_titles) & df['location'].isin(top_locations)]

fig = px.histogram(
    filtered_df,
    x="job_name",
    color="location",
    barmode="group",
    title="Phân phối Job Title theo Location (Top 10)",
    labels={"job_name": "Job Title", "location": "Location"},
    height=700
)

# Apply log scale to y-axis
fig.update_layout(yaxis_type="log")
fig.show()

# 2. Job Title vs Job Type
if 'job_type' in df.columns:
    filtered_df = df[df['job_name'].isin(top_job_titles)]

    fig = px.histogram(
        filtered_df,
        x="job_name",
        color="job_type",
        barmode="group",
        title="Phân phối Job Title theo Job Type",
        labels={"job_name": "Job Title", "job_type": "Job Type"},
        height=700
    )

    # Apply log scale to y-axis
    fig.update_layout(yaxis_type="log")
    fig.show()
```
