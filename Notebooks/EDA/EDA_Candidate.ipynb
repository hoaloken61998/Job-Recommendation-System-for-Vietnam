# Descriptive Analysis

## Univariate Analysis

```python
#Gender
gender_counts = df['gender'].value_counts()
fig_gender = px.pie(
    names=gender_counts.index,
    values=gender_counts.values,
    title='Gender Distribution',
    hole=0.3
)
fig_gender.update_traces(textposition='inside', textinfo='percent+label')

# Increase legend font size
fig_gender.update_layout(
    legend=dict(
        font=dict(
            size=16  # Adjust this number as needed
        )
    )
)

fig_gender.show()
```

```python
# Filter invalid ages
df_cleaned = df[(df['age'] >= 15) & (df['age'] <= 100)]

# Create a cleaner and more polished histogram with box plot
fig_age = px.histogram(
    df_cleaned.dropna(subset=['age']),
    x='age',
    nbins=30,
    marginal="box",          # Adds box plot above
    title='üìä Age Distribution of Candidates',
    color_discrete_sequence=['#636EFA'],  # Set a pleasant color
    opacity=0.75
)

# Update layout for better readability
fig_age.update_layout(
    xaxis_title="Age",
    yaxis_title="Count",
    title_font=dict(size=20, family='Arial'),
    xaxis=dict(tickmode='linear', tick0=15, dtick=5),  # Show age ticks every 5 years
    bargap=0.1,
    plot_bgcolor='rgba(245, 245, 245, 1)',
)

# Optional: boxplot style tweaks
fig_age.update_traces(marker_line_color='black', marker_line_width=0.5)

fig_age.show()
```

```python
# Count frequencies
salary_counts = df['desired_salary'].value_counts()
experience_counts = df['work_experience'].value_counts()

# Create subplot with 1 row and 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=("Desired Salary Distribution", "Work Experience Distribution"))

# Colors from 'husl' (manually define since seaborn colors don't map 1:1 in plotly)
colors_salary = px.colors.qualitative.Plotly[:len(salary_counts)]
colors_experience = px.colors.qualitative.Plotly[:len(experience_counts)]

# Plot 1: Desired Salary
fig.add_trace(
    go.Bar(
        x=salary_counts.index.astype(str),
        y=salary_counts.values,
        marker_color=colors_salary,
        text=salary_counts.values,
        textposition="outside"
    ),
    row=1, col=1
)

# Plot 2: Work Experience
fig.add_trace(
    go.Bar(
        x=experience_counts.index.astype(str),
        y=experience_counts.values,
        marker_color=colors_experience,
        text=experience_counts.values,
        textposition="outside"
    ),
    row=1, col=2
)

# Update layout
fig.update_layout(
    height=500,
    width=1400,
    title_text="Distributions of Salary and Work Experience",
    showlegend=False,
    margin=dict(t=70),
)

# Set log scale for both y-axes
fig.update_yaxes(type="log", title_text="Count", row=1, col=1)
fig.update_yaxes(type="log", title_text="Count", row=1, col=2)

# Set x-axis labels
fig.update_xaxes(title_text="Salary Categories", row=1, col=1)
fig.update_xaxes(title_text="Work Experience (Years)", row=1, col=2)

fig.show()
```

```python
industry_counts = (
    df.groupby('industry')['userid']
    .count()
    .nlargest(10)
    .sort_values(ascending=True)  # Sorted for horizontal bar
)

fig_industry = px.bar(
    x=industry_counts.values,
    y=industry_counts.index,
    orientation='h',
    title='Top 10 Industries by Number of Candidates',
    labels={'x': 'Number of Candidates', 'y': 'Industry'},
    color=industry_counts.values,  # Optional: adds color gradient
    color_continuous_scale='Viridis'
)
fig_industry.update_layout(
    title_font_size=18,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    plot_bgcolor='rgba(0,0,0,0)',
    showlegend=False
)
fig_industry.show()

# 6. Top 10 Desired Workplaces (Vertical Bar Chart)
workplace_counts = (
    df.groupby('workplace_desired')['userid']
    .count()
    .nlargest(10)
    .sort_values(ascending=False)
)

# Create bar chart with categorical colors
fig_workplace = px.bar(
    x=workplace_counts.index,
    y=workplace_counts.values,
    color=workplace_counts.index,  # <- use category for color
    title='Top 10 Desired Workplaces by Number of Candidates',
    labels={
        'x': 'Desired Workplace',
        'y': 'Number of Candidates',
        'color': 'Desired Workplace'
    },
    color_discrete_sequence=px.colors.qualitative.Set2  # or Set1, Pastel, Bold, etc.
)

# Style layout
fig_workplace.update_layout(
    title_font_size=18,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    xaxis_tickangle=45,
    plot_bgcolor='rgba(0,0,0,0)',
    showlegend=False  # Set to True if you want legend
)
```

## Bivariate Analysis

```python
df_exp_salary = df.groupby(['work_experience', 'desired_salary']).size().reset_index(name='count')

fig_exp_salary = px.bar(
    df_exp_salary,
    x='work_experience',
    y='count',
    color='desired_salary',
    barmode='group',
    title='Desired Salary by Work Experience',
)

# Apply log scale to y-axis
fig_exp_salary.update_layout(
    yaxis_type='log',
    yaxis_title='Count',
    xaxis_title='Work Experience',
    title_font=dict(size=20, family='Arial'),
    plot_bgcolor='rgba(245, 245, 245, 1)',
    bargap=0.2
)

fig_exp_salary.show()
```

```python
fig_age_exp = px.box(
    df_cleaned ,
    x='work_experience',
    y='age',
    color='work_experience',
    title='Age Distribution across Work Experience Levels',
    labels={'age': 'Age', 'work_experience': 'Work Experience'},
)
fig_age_exp.show()
```

```python
# Step 1: Get top 5 industries
top_5_industries = df['industry'].value_counts().nlargest(5).index
df_top_ind = df[df['industry'].isin(top_5_industries)]

# Step 2: Count occurrences by industry & workplace
df_ind_workplace = (
    df_top_ind
    .groupby(['industry', 'workplace_desired'])
    .size()
    .reset_index(name='count')
)

# Step 3: Rank workplaces within each industry
df_ind_workplace['rank'] = (
    df_ind_workplace
    .groupby('industry')['count']
    .rank(method='first', ascending=False)
)

# Step 4: Filter top 5 workplaces per industry
df_top_workplaces = df_ind_workplace[df_ind_workplace['rank'] <= 5]

# Step 5: Plot
fig_ind_workplace = px.bar(
    df_top_workplaces,
    x="industry",
    y="count",
    color="workplace_desired",
    title="Top 5 Desired Workplaces per Top 5 Industries",
    labels={'count': 'Number of Candidates', 'industry': 'Industry'},
    barmode='group'
)
fig_ind_workplace.update_layout(
    title_font=dict(size=20),
    xaxis_title="Industry",
    yaxis_title="Number of Candidates"
)

fig_ind_workplace.show()
```
